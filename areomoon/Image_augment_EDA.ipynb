```python
import numpy as np
import matplotlib.pyplot as plt
from dataset import ImageExpDataset
```

```python
dataset = ImageExpDataset(
    fold_file='../AIMango_sample/train_folds.csv',
    pkl_file_path='../AIMango_sample/pkl_files',
    folds=[0,1],
    image_height=224,
    image_width=224,
    mean=(0.485,0.456,0.406),
    std=(0.229,0.224,0.225)
)
```

```python
ind = 20
fig, ax = plt.subplots(2,5,figsize=(20, 6))
for i, axi in enumerate(ax.flat):
    img = dataset[i+ind]['image']
    label = dataset[i+ind]['label']
    np_img = img.numpy()
    axi.imshow(np.transpose(np_img, (1,2,0)))
    class_map = {0:'A',1:'B',2:'C'}
    axi.set_title(f'Grade: {class_map[label.numpy().item(0)]}')
plt.show()
```
